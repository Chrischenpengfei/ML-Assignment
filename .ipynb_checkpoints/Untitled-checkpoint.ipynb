```python
import matplotlib.pyplot as plt
import numpy as np

e = np.e
G = np.arange(0.01,100.0,0.01)
a = 0.01
p = 0.00001

pure_aloha = G * (e ** ((-2 * G)))

slot_aloha = G * (e ** (-G))

#non persistent csma
non_p_csma = (G * (e ** (-a*G)))/(G * (1 + 2 * a) + (e ** (-a*G)))

#one persistent csma
one_p_numer = G * (1 + G + a*G*(1+G+a*G/2))*(e**(-G*(1+2*a)))
one_p_denomin = G * (1 + 2*a) - (1-e**(-a*G))+(1+a*G)*(e**(-G*(1+a)))
one_p_csma = one_p_numer /one_p_denomin

#p persistent csma
pi0 = e**(-G*(1+a))
q = 1 - p
g = a * G
t = (pi0 ** p - pi0) / (1 - pi0 - (pi0 ** p - pi0)*(e ** (-p*g)))
p_s = ((pi0 ** p - pi0) / q * (1 - pi0)) - ((1 - e ** (-g * p)) * (pi0 ** (1 - q ** 2) - pi0)) / (q * (1 - pi0) - (q * (e ** (-2 * g * p)) * (pi0 ** p - pi0)))
p_p_csma1 = (p_s + (1 - pi0) / pi0 * p_s) / (a * t + a * t * ((1 - pi0)/pi0) + (1 + a) / pi0 + a / ( 1 - (e ** (-g))))
p_p_csma2 = ((1 - (e ** (-a*G)))*(p_s*pi0 + p_s*(1 - pi0))) / ((1 - e ** (-a*G)) * (a*t*pi0 + a*t*(1 - pi0) + 1 + a) + a * pi0)

print(pure_aloha.max())
print(slot_aloha.max())
print(non_p_csma.max())
print(one_p_csma.max())
print(p_p_csma1.max())
print(p_p_csma2.max())

plt.figure(num=None, figsize=(15, 10), dpi=100, facecolor='w', edgecolor='k')
plt.xscale('log')
plt.ylim(0, 1)

plt.plot(G, pure_aloha, label = "pure aloha")
plt.plot(G, slot_aloha, label = "slotted aloha")
plt.plot(G, non_p_csma, label = "non persistent CSMA")
plt.plot(G, one_p_csma, label = "one p-persistent CSMA")
plt.plot(G, p_p_csma1, label = "p-persistent CSMA")

plt.grid(alpha=.4,linestyle='--')
plt.legend()
plt.show()
```
